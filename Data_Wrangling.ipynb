### 1. treatments: auralin, novodera: 3 peice of info stored := convert it to 3 cols: lower_dose,upper_dose, treatement_type(auralin, novoders)

```python
#define: a. convert it to 2 cols: dose,treatment_type(auralin, novoders) := solving using melt()

#code:
treatments=pd.melt(treatments, id_vars=['given_name','surname','hba1c_start','hba1c_end','hba1c_change'],var_name='treatment', value_name='dose')

#  id_vars: pass cols in which no change is requred
#  varname: combines remaning col into single col with column name as value
#  value_name: stores value of col
```

```python
treatments.head()
```

```python
#  patient where auaralin dose is '-' means -> novoders was given to them and vice versa

#  remove those
treatments=treatments[treatments['dose']!='-']
```

```python
# define: b. seperate dose into lower dose, upper dose
treatments['start_dose'],treatments['end_dose']=treatments['dose'].str.split('-').str
```

```python
treatments.head()
```

```python
# remove redundent dose column
treatments.drop(columns={'dose'}, inplace=True)
```

```python
treatments
```

```python
#reseting index
treatments.reset_index(drop=True, inplace=True)
```

```python
treatments
```

```python
treatments.info()
```

### 2. adverse effect: no need of seperate table

```python
#merging 
patients=patients.merge(adverse, how='left', left_on=['given_name','surname'], right_on = ['given_name','surname'])
```

## 3. Remaning Data quality issues (Validity, accuracy and consistency)

### Validity

1. patient
	- zip code: in decimal [convert to string] and some zip code of 4 digit
2. treatment  
	- novodra, auralin: u attached with numbers 
	- novodra, auralin: inconsistent data type: convert object to neumarical data
 


```python
## 1. patient:  zip code: in decimal [convert to string]   

## removing null zipcodes
patients=patients[patients['zip_code'].isnull()==False]
```

```python
#zip code: in decimal [float to int]   
patients['zip_code']=patients['zip_code'].astype(int)
```

```python
# int to string
patients['zip_code']=patients['zip_code'].astype(str)
```

```python
# convert xip to 5 length
patients['zip_code']=patients['zip_code'].str.zfill(5)
```

```python
## 2. treatment:   novodra, auralin: u attached with numbers 
treatments['start_dose']=treatments['start_dose'].str.replace('u','')
treatments['end_dose']=treatments['end_dose'].str.replace('u','')
treatments.head()
```

```python
## 2. treatment novodra, auralin: inconsistent data type: convert object to neumarical data

#checking type
print(type(treatments['end_dose'][0]),type(treatments['start_dose'][0]))
```

```python
#converting to int
treatments['start_dose']=treatments['start_dose'].astype(int)
treatments['end_dose']=treatments['end_dose'].astype(int)
```
